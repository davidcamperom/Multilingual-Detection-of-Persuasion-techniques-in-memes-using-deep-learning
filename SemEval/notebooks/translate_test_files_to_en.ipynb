```python
# Importar librerías
import json
import os
from easynmt import EasyNMT
import pandas as pd
import random

# set up root path
root_path = os.getcwd()

# path to the datasets
data_path = f'{root_path}/data/scorer-baseline/'
```

# Arabic data

```python
ar_data = f'{data_path}arabic/ar_subtask1_majority.txt'
with open(ar_data,'r') as f:
    data = json.load(f)
data[0]
```

```python
ar_data_translated = []
translation_model = EasyNMT('opus-mt')

for d in data:
    text = d['text']
    translated_text = translation_model.translate(text,source_lang='ar', target_lang='en')
    print(translated_text)
    d['text'] = translated_text
    ar_data_translated.append(d)
```

```python
ar_data_translated
```

```python
with open(f'{data_path}arabic/ar_subtask1_majority_en_translated.txt', 'w') as f:
    json.dump(ar_data_translated, f, indent=4)
```
